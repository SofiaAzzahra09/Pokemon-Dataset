Visualisasi Data

```python
# Distribusi total statistik dari 'Total'
plt.figure(figsize=(10, 6))
sns.histplot(data=new_data, x='Total', bins=20)
plt.title('Distribusi Total Statistik Pokemon')
plt.xlabel('Total Statistik')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Mengetahui jumlah pokemon dari tipe 1
plt.figure(figsize=(12, 6))
sns.countplot(data=new_data, x='Type 1', order=new_data['Type 1'].value_counts().index)
plt.title('Jumlah Pokemon Berdasarkan Jenis (Type 1)')
plt.xlabel('Jenis Pokemon')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()
```

```python
# Mengetahui jumlah pokemon dari tipe 2
plt.figure(figsize=(12, 6))
sns.countplot(data=new_data, x='Type 2', order=new_data['Type 2'].value_counts().index)
plt.title('Jumlah Pokemon Berdasarkan Jenis (Type 2)')
plt.xlabel('Jenis Pokemon')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.show()
```

```python
# Menunjukkan hubungan antara serangan dan pertahanan
plt.figure(figsize=(8, 6))
sns.scatterplot(data=new_data, x='Attack', y='Defense', hue='Legendary')
plt.title('Hubungan Antara Serangan dan Pertahanan Pokemon')
plt.xlabel('Serangan')
plt.ylabel('Pertahanan')
plt.show()
```

```python
# Jumlah Pokemon legendaris berdasarkan tipe 1
plt.figure(figsize=(12, 6))
sns.countplot(data=new_data, x='Type 1', hue='Legendary')
plt.title('Jumlah Pokemon Legendaris dan Non-legendaris Berdasarkan Tipe 1')
plt.xlabel('Tipe')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.legend(title='Legendary', loc='upper right')
plt.show()
```

```python
# Jumlah Pokemon legendaris berdasarkan tipe 2
plt.figure(figsize=(12, 6))
sns.countplot(data=new_data, x='Type 2', hue='Legendary')
plt.title('Jumlah Pokemon Legendaris dan Non-legendaris Berdasarkan Tipe 2')
plt.xlabel('Tipe')
plt.ylabel('Jumlah')
plt.xticks(rotation=45)
plt.legend(title='Legendary', loc='upper right')
plt.show()
```
